```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/tvmarketing.csv')
```

```python
df.describe()
```

```python
X = df.iloc[:,[0]]
X
```

```python
y = df.iloc[:,[1]]
y
```

```python
# Splitting the dataset into training and test set.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size= 1/3, random_state=0)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
y_train.head()
```

```python
y_test.head()
```

```python
#Fitting the Simple Linear Regression model to the training dataset
from sklearn.linear_model import LinearRegression
regressor= LinearRegression()
regressor.fit(X_train, y_train)
```

```python
#Prediction of Test and Training set result
y_pred= regressor.predict(X_test)
X_pred= regressor.predict(X_train)
```

```python
# Training set
plt.scatter(X_train, y_train, color="green")
plt.plot(X_train, X_pred, color="red")
plt.title("TV vs Sales (Training Data Visualization)")
plt.xlabel("TV Price")
plt.ylabel("Sales")
plt.show()
```

```python
#visualizing the Test set results
plt.scatter(X_test, y_test, color="blue")
plt.plot(X_train, X_pred, color="red")
plt.title("TV vs Sales (Test Data Visualization)")
plt.xlabel("TV Price")
plt.ylabel("Sales")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Calculate Mean Squared Error for both training and test sets
mse_train = mean_squared_error(y_train, X_pred)
mse_test = mean_squared_error(y_test, y_pred)

print("Mean Squared Error (Training Set):", mse_train)
print("Mean Squared Error (Test Set):", mse_test)
```

```python
# Calculate R-squared for both training and test sets
r2_train = r2_score(y_train, X_pred)
r2_test = r2_score(y_test, y_pred)

print("R-squared (Training Set):", r2_train*100)
print("R-squared (Test Set):", r2_test*100)
```
