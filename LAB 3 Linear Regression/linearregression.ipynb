```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as mlt
```

```python
df = pd.read_csv("CAR DETAILS FROM CAR DEKHO.csv")
```

```python
df1 = df[['km_driven', 'selling_price']]
```

```python
df1
```

```python
X = df1.iloc[:, :-1].values  
y = df1.iloc[:, 1].values 
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size= 1/3, random_state=0) 
```

```python
from sklearn.linear_model import LinearRegression  
regressor= LinearRegression() 
regressor.fit(X_train, y_train)
```

```python
#Prediction of Test and Training set result  
y_pred= regressor.predict(X_test)  
X_pred= regressor.predict(X_train)  
```

```python
mlt.scatter(X_train, y_train, color="green")   
mlt.plot(X_train, X_pred, color="red")    
mlt.title("KMs driven vs Price (Training Dataset)")  
mlt.xlabel("KMs driven")  
mlt.ylabel("Selling Price(In Rupees)")  
mlt.show()   
```

```python
mlt.scatter(X_test, y_test, color="green")   
mlt.plot(X_test, X_pred, color="red")    
mlt.title("KMs driven vs Price (Training Dataset)")  
mlt.xlabel("KMs driven")  
mlt.ylabel("Selling Price(In Rupees)")  
mlt.show()   
```
